```python
import numpy as np
rng = np.random.RandomState(0)

x = rng.randint(1, 100, 50)
y = rng.randint(1, 100, 50)

plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Student score - Maths vs Physics')

plt.scatter(x, y)
plt.show()
```

```python
import numpy as np
rng = np.random.RandomState(0)

x = rng.randint(1, 100, 50)
y = rng.randint(1, 100, 50)

plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Student score - Maths vs Physics')

plt.scatter(x, y, s=100, color='green', marker='v') # s is size of scattered dots: v made v-shaped
plt.show()
```

```python
import numpy as np
rng = np.random.RandomState(0)

x = rng.randint(1, 100, 50)
y = rng.randint(1, 100, 50)

colors = rng.randint(1, 100, 50)
sizes = 2 * np.add(x, y) 

plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Student score - Maths vs Physics')

plt.scatter(x, y, c=colors, s=sizes, alpha = 0.5, cmap='cividis') # alpha do the transparency
plt.colorbar() 
plt.show()
```
