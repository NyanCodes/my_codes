```python
import matplotlib.pyplot as plt

x = ['Samsung', 'Nokia', 'Apple', 'Oppo', 'Huawei', 'Sony']
y = [22 , 14, 13, 32, 19, 11]

plt.bar(x, y)
plt.show()
```

```python
plt.ylabel('Usage')
plt.xlabel('Mobile Phone')
plt.title('Usage of Mobile Phone')
plt.bar(x, y)
plt.show()
```

```python
import numpy as np
labels = ['Samsung', 'Nokia', 'Apple', 'Oppo', 'Huawei', 'Sony']
men = [10, 3, 9, 10, 2, 10]
women = [10, 2, 10, 9, 3, 10]

fig, ax = plt.subplots()
width = 0.4

y=np.arange(6)

ax.bar(y-width/2, men, width, label="Man")
ax.bar(y+width/2, women, width, label="Woman")
ax.legend()
ax.set_xticks(x)
ax.set_xticklabels(labels)
plt.show()
```

```python
plt.bar(labels, men, color='red')
plt.bar(labels, women, bottom=men, color='green')

plt.xlabel('Brand')
plt.ylabel('Usage')
plt.show()
```
