```python
import matplotlib.pyplot as plt

x1 = [40, 40, 80, 80, 80, 80, 80, 75, 75, 75, 50, 50, 50, 50, 60, 90, 35, 35, 32]
x2 = [20, 20, 14, 14, 10, 10, 10, 10, 10, 90, 80, 90, 30, 30, 40, 40, 50, 55, 60]

num_bins = 7

plt.hist(x1, num_bins, label='Maths', alpha=0.5, color='blue')
plt.hist(x2, num_bins, label='Physics', alpha=0.5, color='red')
plt.xlabel('Marks')
plt.ylabel('Physics')
plt.title('Student Scores')
plt.legend(loc='upper right')
plt.show()
```
