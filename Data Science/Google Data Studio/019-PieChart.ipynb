```python
%matplotlib inline
import matplotlib.pyplot as plt

data = [100, 50, 150]

lbl = ['Coffee', 'Tea', 'Milk']
col = ['brown', 'orange', 'gray']
spl = [0, 0.1, 0] # explode

plt.pie(data, labels = lbl, colors = col, explode = spl, shadow = True, autopct = '%1.2f%%') 
                                        # autopct show the percentage
plt.title('Drink Distribution')

plt.axis('equal')
plt.show()
```
