```python
import matplotlib.pyplot as plt

x = [1, 3 , 5]
y = [1, 3 , 5]
plt.plot(x, y)
plt.show()
```

```python
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Chart Example')
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, linestyle='dotted', linewidth=2.5)
```

```python
import numpy as np 

x = np.linspace(1, 10, 50)
print(x)
```

```python
y = 2 ** x
```

```python
plt.plot(x, y, label='Creating exponential line chart')
plt.legend()
plt.title('Hi')
```
